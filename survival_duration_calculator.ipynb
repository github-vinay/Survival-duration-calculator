```python
age = float(input("What's your age? "))
unit = input("Choose a time unit: Months (m), Weeks (w), Days (d), Hours (h), Minutes (mi), Seconds (s): ")

if age < 0:
    print("You are not born yet!")
elif age <= 1:
    print("Newborn")
    if unit == "m":
        print(f"You lived for {age * 12} months.")
    elif unit == "w":
        print(f"You lived for {age * 52} weeks.")
    elif unit == "d":
        print(f"You lived for {age * 365} days.")
    elif unit == "h":
        print(f"You lived for {age * 365 * 24} hours.")
    elif unit == "mi":
        print(f"You lived for {age * 365 * 24 * 60} minutes.")
    elif unit == "s":
        print(f"You lived for {age * 365 * 24 * 60 * 60} seconds.")
    else:
        print("Invalid time unit.")
else:
    if unit == "m":
        print(f"You lived for {age * 12} months.")
    elif unit == "w":
        print(f"You lived for {age * 52} weeks.")
    elif unit == "d":
        print(f"You lived for {age * 365} days.")
    elif unit == "h":
        print(f"You lived for {age * 365 * 24} hours.")
    elif unit == "mi":
        print(f"You lived for {age * 365 * 24 * 60} minutes.")
    elif unit == "s":
        print(f"You lived for {age * 365 * 24 * 60 * 60} seconds.")
    else:
        print("Invalid time unit.")
```
